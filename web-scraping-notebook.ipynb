```python
# import

import requests
import io
from bs4 import BeautifulSoup as BS
import pandas as pd
```

# 1. Start by performing a GET request on the url above and convert the response into a BeautifulSoup object.
### a. Use the .find method to find the tag containing the first job title ("Senior Python Developer"). Hint: can you find a tag type and/or a class that could be helpful for extracting this information? Extract the text from this title.


```python
# get em
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url, headers=headers)

# check response
print('the response code is ' + str(response.status_code))
```

```python
# convert to BS
soup = BS(response.text)

# sample the BS
soup
```

```python
#first h2 tag with class title on the page
first_job_title = soup.find_all('h2', attrs={'class':'title'})[0].text
first_job_title
```

# b. Now, use what you did for the first title, but extract the job title for all jobs on this page. Store the results in a list.

```python
# Find all h2 tags with class 'title'
job_titles = soup.find_all('h2', class_='title')

# Get the text from each h2 tag
job_title_list = [x.get_text() for x in job_titles]

# Check the type (optional)
job_title_list
```

# c. Finally, extract the companies, locations, and posting dates for each job. For example, the first job has a company of "Payne, Roberts and Davis", a location of "Stewartbury, AA", and a posting date of "2021-04-08". Ensure that the text that you extract is clean, meaning no extra spaces or other characters at the beginning or end.


```python
# Find companies
job_companies = soup.find_all('h3', class_='subtitle is-6 company')

# now a list
job_companies_list = [x.get_text() for x in job_companies]

# Find locations
job_locs = soup.find_all('p', class_='location')

# now a list with linebreaks and whitespace stripped
job_locs_list = [x.get_text(strip=True) for x in job_locs]

#Find posting dates
job_dates = soup.find_all('time')
job_dates_list = [x.get_text() for x in job_dates]

# test it
# job_companies_list
# job_locs_list
# job_dates_list
```

# d. Take the lists that you have created and combine them into a pandas DataFrame.

```python
#make dictionary out of the lists
job_dict = {'company':job_companies_list,
            'location':job_locs_list,
            'posting_date':job_dates_list}

df = pd.DataFrame(job_dict)
df
```
